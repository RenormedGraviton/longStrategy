```python
def stable_scale(arr, scale):
    return (1.0 - np.exp(-1.0*abs(arr)/scale))*np.sign(arr)
```

```python
scale_profiles= np.absolute(np.array([stable_scale(profiles_valid[i], 3.0) for i in range(profiles_valid.shape[0])]))
scale_profiles_norm = scale_profiles - scale_profiles[:,:1]
```

```python
plt.figure()
plt.title('abs(cvd_usdt)')
for i in range(0, len(profiles_valid)):
    plt.plot(scale_profiles_norm[i], color = 'grey', alpha = 0.2)
plt.axvline(24*20, color = 'red', linestyle = 'dashed')
plt.plot(np.median(scale_profiles_norm, axis = 0), color = 'red')
plt.grid(True)
```
